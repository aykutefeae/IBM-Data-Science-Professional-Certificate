## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
box_ConvComp = df.boxplot(column=["ConvertedComp"], grid=False, figsize=(12,8))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
df["ConvertedComp"].describe()
```

```python
df["ConvertedComp"].median()
```

```python
# your code goes here
df["ConvertedComp"].dropna(axis=0, inplace=True)
Q1_ConvComp = df["ConvertedComp"].quantile(.25)
Q3_ConvComp = df["ConvertedComp"].quantile(.75)
IQR_ConvComp = Q3_ConvComp - Q1_ConvComp
IQR_ConvComp
```

Find out the upper and lower bounds.


```python
# your code goes here
upper_ConvComp = Q3_ConvComp + (IQR_ConvComp*1.5)
lower_ConvComp = Q1_ConvComp - (IQR_ConvComp*1.5)

print(upper_ConvComp)
print(lower_ConvComp)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers_ConvComp = (df["ConvertedComp"]<lower_ConvComp) | (df["ConvertedComp"]>upper_ConvComp)
outliers_ConvComp.value_counts()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df[~((df["ConvertedComp"] < lower_ConvComp) | (df["ConvertedComp"] > upper_ConvComp))]
```

```python
df2
```

```python
df2["ConvertedComp"].describe()
```

```python
df2["ConvertedComp"].median()
```
